```python
import pandas as pd
import datetime
import numpy as np
from datetime import timedelta
test_data = pd.read_csv('test_data2.csv',encoding='utf-8-sig')
#format date_of_trade
test_data['date_of_trade'] = pd.to_datetime(test_data['date_of_trade'],format='%Y/%m/%d')
#sort by date_of_trade

test_data = test_data.sort_values(by=['date_of_trade'])
```

```python
test_data
```

```python
#seperate test data into 2 parts by the column "type_of_trade"
purchase_data = test_data[test_data['type_of_trade'] == "purchase"]
sale_data = test_data[test_data['type_of_trade'] == "sale"]
```

```python
#get the smallest date in test_data

date_min = test_data['date_of_trade'].min()
date_max = test_data['date_of_trade'].max()

#calculate the date range
if date_max - date_min <= timedelta(days=180):
    print("The date range is less than 180 days")
else:
    print("The date range is more than 180 days")

#calculate the date range
date_range = date_max - date_min

print("The date range is " + str(date_range.days) + " days")
```

```python
def time_check(highest_sale,lowest_purchase):
    highest_sale['date_of_trade'] = pd.to_datetime(highest_sale['date_of_trade'],format='%Y/%m/%d')
    lowest_purchase['date_of_trade'] = pd.to_datetime(lowest_purchase['date_of_trade'],format='%Y/%m/%d')
    date_range = highest_sale['date_of_trade'] - lowest_purchase['date_of_trade'] 
    if date_range <= timedelta(days=180):
        time_check = False
    else:
        time_check = True
    return time_check
    
```

```python
profit = 0
current_profit = 1
while current_profit > 0 :
    # find highest sale price and lowest purchase price
    highest_sale = sale_data.loc[sale_data['sale_price'].idxmax()]
    lowest_purchase = purchase_data.loc[purchase_data['purchase_price'].idxmin()]
    
    if highest_sale['sale_share'] > lowest_purchase['purchase_share']:
        c_share = highest_sale['sale_share'] - lowest_purchase['purchase_share']
        current_profit = highest_sale['sale_price'] * lowest_purchase['purchase_share'] - lowest_purchase['purchase_price'] * lowest_purchase['purchase_share']
        print("current_profit:" +str(current_profit))
        print("highest_sale['sale_price']:" +str(highest_sale['sale_price']))
        print("highest_sale['sale_share']:" +str(highest_sale['sale_share']))
        print("lowest_purchase['purchase_price']:" +str(lowest_purchase['purchase_price']))
        print("lowest_purchase['purchase_share']:" +str(lowest_purchase['purchase_share']))
        if current_profit < 0:
            print("break")
            break
        else:
            print("A:" +str(current_profit))
            profit += current_profit
            #update sale_data
            sale_data.loc[sale_data['sale_price'].idxmax(),'sale_share'] = c_share
            #drop lowest_purchase from purchase_data
            #purchase_data = purchase_data.drop(purchase_data.index[lowest_purchase.name])
            purchase_data.loc[purchase_data['purchase_price'].idxmin(),'purchase_share'] = np.nan
            purchase_data.loc[purchase_data['purchase_price'].idxmin(),'purchase_price'] = np.nan

            
                
                
    elif highest_sale['sale_share'] < lowest_purchase['purchase_share']:
        c_share = lowest_purchase['purchase_share'] - highest_sale['sale_share']
        current_profit = highest_sale['sale_price'] * highest_sale['sale_share'] - lowest_purchase['purchase_price'] * highest_sale['sale_share']
        print("current_profit:" +str(current_profit))
        print("highest_sale['sale_price']:" +str(highest_sale['sale_price']))
        print("highest_sale['sale_share']:" +str(highest_sale['sale_share']))
        print("lowest_purchase['purchase_price']:" +str(lowest_purchase['purchase_price']))
        print("lowest_purchase['purchase_share']:" +str(lowest_purchase['purchase_share']))
        if current_profit < 0:
            print("break")
            break
        else:
            print("B:" +str(current_profit))
            profit += current_profit
            #update purchase_data
            purchase_data.loc[purchase_data['purchase_price'].idxmin(),'purchase_share'] = c_share
            #update highest_sale from sale_data
            sale_data.loc[sale_data['sale_price'].idxmax(),'sale_share'] = np.nan
            sale_data.loc[sale_data['sale_price'].idxmax(),'sale_price'] = np.nan
            
                
                

    else:
        current_profit = highest_sale['sale_price'] * highest_sale['sale_share'] - lowest_purchase['purchase_price'] * lowest_purchase['purchase_share']
        print("current_profit:" +str(current_profit))
        print("highest_sale['sale_price']:" +str(highest_sale['sale_price']))
        print("highest_sale['sale_share']:" +str(highest_sale['sale_share']))
        print("lowest_purchase['purchase_price']:" +str(lowest_purchase['purchase_price']))
        print("lowest_purchase['purchase_share']:" +str(lowest_purchase['purchase_share']))
        if current_profit < 0:
            print("break")
            
        else:
            print("C:" +str(current_profit))
            
            profit += current_profit
            #update sale_data
            sale_data.loc[sale_data['sale_price'].idxmax(),'sale_share'] = np.nan
            sale_data.loc[sale_data['sale_price'].idxmax(),'sale_price'] = np.nan
            #update purchase_data 
            purchase_data.loc[purchase_data['purchase_price'].idxmin(),'purchase_share'] = np.nan
            purchase_data.loc[purchase_data['purchase_price'].idxmin(),'purchase_price'] = np.nan



print("依據證券交易法§157短線交易歸入權，應歸入的金額為"+str(profit))

                
            
        

        
        


        
```
